```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression

data = {
    "area": [1000, 1500, 1800, 2000],
    "bedrooms": [2, 3, 3, 4],
    "price": [50, 75, 90, 120]
}

df = pd.DataFrame(data)

X = df[["area", "bedrooms"]]
y = df["price"]

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[1700, 3]])

print("Predicted Price:", prediction)
```
